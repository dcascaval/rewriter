### Define Shape Program DSL

```python
import numpy as np
from dataclasses import dataclass
from typing import List
from math import sqrt

@dataclass
class Number:
    val: float
    
num = Number

def box_union(*boxes):
    points = np.concatenate(boxes, 0)
    return np.stack([points.min(axis=0), points.max(axis=0)])

class Shape:
    pass

@dataclass
class Point(Shape):
    x: num
    y: num

    def eval(self):
        return f'<circle cx="{self.x}" cy="{self.y}" r="1" fill="green"/>'

    def box(self):
        return np.array([[self.x,self.y],[self.x,self.y]])

    def get_params(self):
        return [self.x, self.y]

    def __add__(self, other):
        return Point(self.x+other.x, self.y+other.y)
    
    def __sub__(self, other):
        return Point(self.x-other.x, self.y-other.y)
    
    def __mul__(self, s):
        return Point(s*self.x, s*self.y)

    def __rmul__(self, s):
        return self.__mul__(s)

@dataclass
class Line(Shape):
    start: Point
    end: Point

    def eval(self):
        return f'<line x1="{self.start.x}" y1="{self.start.y}" x2="{self.end.x}" y2="{self.end.y}" stroke-width="1" stroke="red"/>'
    
    def length(self):
        return sqrt((self.end.x - self.start.x)**2 + (self.end.y - self.start.y)**2)

    def box(self):
        return box_union(self.start.box(), self.end.box())

    def get_params(self):
        return self.start.get_params() + self.end.get_params()

def Evaluate(line, position):
    x = line.start.x
    y = line.start.y
    dx = (line.end.x - x)
    dy = (line.end.y - y)
    return Point(x+position*dx, y+position*dy)

@dataclass
class Extrude(Shape):
    line: Line
    distance: num

    def poly(self):
        norm = self.line.length()
        perp_y = self.distance * (self.line.end.x - self.line.start.x) / norm
        perp_x = self.distance * (self.line.start.y - self.line.end.y) / norm
        rect = Polygon([
           self.line.start,
           self.line.end,
           Point(self.line.end.x + perp_x, self.line.end.y + perp_y),
           Point(self.line.start.x + perp_x, self.line.start.y + perp_y)
        ])

        return rect

    def eval(self):
        return self.poly().eval()

    def box(self):
        return self.poly().box()

    def get_params(self):
        return 

@dataclass
class Polygon(Shape):
    points: List[Point]

    def eval(self):
        points_str = " ".join(f'{p.x},{20-p.y}' for p in self.points)
        return f'<polygon points="{points_str}" fill="grey" stroke="black" stroke-width="0.1"/>'

    def box(self):
        return box_union(*[p.box() for p in self.points])

@dataclass
class Rectangle(Shape):
    c1: Point
    c2: Point

    def box(self):
        return box_union(self.c1.box(), self.c2.box())

    def eval(self):
        w = self.c2.x - self.c1.x
        h = self.c2.y - self.c1.y   
        return f'<rect x="{self.c1.x}" y="{20-self.c1.y-h}" width="{w}" height="{h}" fill="grey" stroke="black" stroke-width="0.1"/>'

@dataclass
class Box:
    w: num
    h: num

    def box(self):
        return np.array([[0,0], [self.w, self.h]])
    
    def eval(self):
        return f'<rect width="{self.w}" height="{self.h}" x="0" y="20-{self.h}" fill="grey" stroke="black" stroke-width="0.1"/>'

@dataclass
class ShapeList(Shape):
    shapes: List[Shape]

    def box(self):
        return box_union(*[s.box() for s in self.shapes])
    
    def eval(self):
        contents_svg = '\n'.join([s.eval() for s in self.shapes])
        return f'<g>{contents_svg}</g>'

@dataclass
class Translate(Shape):
    shape: Shape
    x: num
    y: num

    def box(self):
        return self.shape.box() + np.array([self.x, self.y])

    def eval(self):
        return f'<g transform="translate({self.x} {-self.y})">{self.shape.eval()}</g>'

@dataclass
class Scale(Shape):
    shape: Shape
    s: num

    def box(self):
        return self.shape.box()*self.s

    def eval(self):
        return f'<g transform="scale({self.s} {self.s})">{self.shape.eval()}</g>'
```

### Program Examples with Good / Bad Samples

```python
def simple_box(w, h):
    box = Rectangle(Point(0,0),Point(w,h))
    return box.eval()

simple_box_examples = {
    'scale_param': {
        'good': [
            [1,1],
            [2,2]
        ],
        'bad': [
            [1,2],
            [2,1]
        ]
    },
    'height_param': {
        'good': [
            [1,1],
            [1,2]
        ],
        'bad': [
            [2,1],
            [2,2]
        ]
    },
    'width_param': {
        'good': [
            [1,1],
            [2,1]
        ],
        'bad': [
            [1,2],
            [2,2]
        ]
    }
}
```

```python
def extruded_box(x1,y1,x2,y2,d):
    a = Point(x1,y1)
    b = Point(x2,y2)
    l = Line(a, b)
    box = Extrude(l, d)
    return box.eval()

extruded_box_examples = {
    'rotating_box': {
        'good':[
            [10,10,13,14,5],
            [10,10,14,13,5],
            [10,10,15,10,5],
            [10,10,13,14,3],
            [10,10,14,13,6],
            [10,10,15,10,-1],
        ],
        'bad':[

            [8,10,12,14,5],
            [8,10,12,13,5],
            [8,10,13,10,5],
            [10,10,14,14,5],
        ]
    },
    'rotating_square': {
        'good':[
            [10,10,13,14,5],
            [10,10,14,13,5],
            [10,10,15,10,5]
        ],
        'bad':[
            [10,10,13,14,3],
            [10,10,14,13,6],
            [10,10,15,10,-1],
            [8,10,12,14,5],
            [8,10,12,13,5],
            [8,10,13,10,5],
            [10,10,14,14,5],
        ]
    },
    'no_rotation':{
        'good': [
            [10,10,13,13,3],
            [10,10,13,13,6],
            [10,10,15,15,3],
            [5,6,8,9,3],
        ],
        'bad': [
            [10,11,13,13,3],
            [9,10,13,13,6],
        ]
    },
    'no_rot_on_axis':{
        'good': [
            [10,10,13,13,3],
            [10,10,13,13,6],
            [10,10,15,15,3],
            [5,5,8,8,4],
        ],
        'bad': [
            [10,11,13,13,3],
            [9,10,13,13,6],
            [5,6,8,9,3],
        ] 
    }
}
```

```python

def stairs_1(dx, dy, N):
    step = Rectangle(Point(0,0), Point(dx,dy))
    steps = [Translate(step, i*dx, i*dy) for i in range(N)]
    stairs = ShapeList(steps)
    return stairs.eval()

stairs_1_examples = {
    'more_stairs': {
        'good': [
            [2,1,5],
            [2,1,10]
        ],
        'bad': [
            [4,2,5]
        ]
    },
    'big_stairs': {
        'good': [
            [2,1,5],
            [4,2,5]
        ],
        'bad': [
            [2,1,10]
        ]
    }
}
```

```python
def stairs_2(x, y, N):
    l = Line(Point(0,0), Point(x,y))
    corners = [Evaluate(l, i/N) for i in range(N)]
    step = corners[1] - corners[0]
    stairs = ShapeList([Rectangle(corner, corner+step) for corner in corners])
    return stairs.eval()

stairs_2_examples = {
    'more_stairs': {
        'good': [
            [10, 5, 5],
            [20, 10, 10]
        ],
        'bad': [
            [20, 10, 5]
        ]
    },
    'big_stars': {
        'good': [
            [10, 5, 5],
            [20, 10, 5]
        ],
        'bad': [
            [20, 10, 10]
        ]
    }
}
```

### Helper Function to Generate an SVG of an Examples Gallery

```python
def make_examples_figure(program, examples, title, max_cols = None):
    n_rows = 2*len(examples)
    n_cols = max(*[max(len(ex['good']), len(ex['bad'])) for _,ex in examples.items()])
    if max_cols:
        n_cols = min(n_cols, max_cols)
    
    svgs = []

    s = 26

    for row, (name, example) in enumerate(examples.items()):

        good_offset = 2*s*row-2
        bad_offset = s*(2*row+1)-2

        for col, ex in enumerate(example['good']):
            col_offset = s*col-2
            ex_svg = add_grid(program(*ex))
            svgs.append(transform_svg(ex_svg, (col_offset, good_offset)))
            t_x = col_offset + s/2 - len(ex) - 2
            t_y = good_offset - 0.5
            svgs.append(f'<text x="{t_x}" y="{t_y}" font-size="1">{str(ex)}</text>')
            if col > n_cols:
                break

        for col, ex in enumerate(example['bad']):
            col_offset = s*col-2
            ex_svg = add_grid(program(*ex))
            svgs.append(transform_svg(ex_svg, (col_offset, bad_offset)))
            t_x = col_offset + s/2 - len(ex) - 2
            t_y = bad_offset - 0.5
            svgs.append(f'<text x="{t_x}" y="{t_y}" font-size="1">{str(ex)}</text>')
            if col > n_cols:
                break

        gt_x = -5
        gt_y = good_offset + s/2
        bt_x = -5
        bt_y = bad_offset + s/2
        svgs.append(f'<text x="{gt_x}" y="{gt_y}" transform="rotate(-90,{gt_x},{gt_y})" font-size="1">{name}: good</text>')
        svgs.append(f'<text x="{bt_x}" y="{bt_y}" transform="rotate(-90,{bt_x},{bt_y})" font-size="1">{name}: bad</text>')

    svgs.append(f'<text x="-2" y="-6" font-size="1.5">{title}</text>')

    svg = ''
    for sv in svgs:
        svg += sv
    
    return wrap_svg(svg, (-6, -8, s*n_cols + 1, s*(n_rows)+2))
```
